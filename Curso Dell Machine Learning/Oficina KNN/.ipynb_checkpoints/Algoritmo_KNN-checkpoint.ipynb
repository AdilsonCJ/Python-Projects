```python
# selecione no dataframe df as colunas Age e Spending Score (1-100)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importação do Dataset a ser utilizado
df = pd.read_csv('Mall_Customers.csv')
```

```python
# Avaliação Primária da Base de Dados
df.describe()
df.shape
df.isnull().sum()
df.columns
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'],marker='.')
plt.xlabel('Idade dos Clientes')
plt.ylabel('Score do Clientes (1 ~ 100)')
plt.show()
#Interessante Observar : Pessoas com faixa etária acima de 40 anos tem atividade de compra limitada à 60 pontos
```

### Clusterizando com k = 5

```python
# escreva aqui em baixo o código para fazer a clusterização
#Início do Código
number_cluster = 5
X = df[['Age','Spending Score (1-100)']]
modelo_kmeans = KMeans(n_clusters=number_cluster,init='k-means++')
y_kmeans = modelo_kmeans.fit_predict(X)
cores = ['r', 'b', 'k', 'y', 'g']
```

### Plot um gráfico com todos os grupos

```python
# escreva aqui o código para plotar o gráfico
for k in range(number_cluster):
    cluster = X[y_kmeans == k]
    plt.scatter(cluster['Age'], cluster['Spending Score (1-100)'],s = 100, c = cores[k], label = f'Cluster {k}')
                
plt.title('Avaliação Clientes vs Score-Gastos')
plt.xlabel('Idade dos Clientes')
plt.ylabel('Score dos Clientes (1~100)')
plt.grid()
plt.legend()
plt.show()
```

### Clusterizando com k = 6

```python
# escreva aqui em baixo o código para fazer a clusterização
#Início do Código
number_cluster_2 = 6
X_2 = df[['Age','Spending Score (1-100)']]
modelo_kmeans = KMeans(n_clusters=number_cluster_2,init='k-means++')
y_kmeans = modelo_kmeans.fit_predict(X_2)
cores_new = ['r', 'b', 'k', 'y', 'g','m']

# escreva aqui o código para plotar o gráfico
for k in range(number_cluster_2):
    cluster = X_2[y_kmeans == k]
    plt.scatter(cluster['Age'], cluster['Spending Score (1-100)'],s = 100, c = cores_new[k], label = f'Cluster {k}')
                
plt.title(f'Avaliação Clientes vs Score-Gastos K = {number_cluster_2}')
plt.xlabel('Idade dos Clientes')
plt.ylabel('Score dos Clientes (1~100)')
plt.grid()
plt.legend()
plt.show()
```

## Clusterizando com k = 7

```python
# escreva aqui em baixo o código para fazer a clusterização
#Início do Código
number_cluster_3 = 7
X_3 = df[['Age','Spending Score (1-100)']]
modelo_kmeans = KMeans(n_clusters=number_cluster_3,init='k-means++')
y_kmeans = modelo_kmeans.fit_predict(X_3)
cores_new_2 = ['r', 'b', 'k', 'y', 'g','m','c']

# escreva aqui o código para plotar o gráfico
for k in range(number_cluster_3):
    cluster = X_3[y_kmeans == k]
    plt.scatter(cluster['Age'], cluster['Spending Score (1-100)'],s = 100, c = cores_new_2[k], label = f'Cluster {k}')
                
plt.title(f'Avaliação Clientes vs Score-Gastos K = {number_cluster_3}')
plt.xlabel('Idade dos Clientes')
plt.ylabel('Score dos Clientes (1~100)')
plt.grid()
plt.legend()
plt.show()
```
