##### Importe as bibliotecas pandas, matplotlib e seaborn, usando a palavra mágica inline, para exibir as visualizações no notebook. Carregue os dados, que podem ser carregados pelo seaborn, utilizando a função load_dataset(‘mpg’), em um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_car = sns.load_dataset("mpg")
df_car.head()
```

##### Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!

```python
sns.relplot(x="mpg", y="acceleration", col = 'origin', data=df_car)

print('Em um primeiro momento, é curioso ver como os carros Europeus e Japoneses possuem uma faixa mínima de aceleração maior do que os americanos, deixando espaços em branco no chart')
print('\nCom relação à correlação : Embora os três tipos de carros parecem apresentar uma correção próxima de 0 (Ou seja, sem correlação entre mpg e aceleração), podemos talvez enxergar uma correlação baixa positiva nos carros americanos, aonde a aceleração interfere minimamente no consumo de combustível')
```

##### Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

```python
sns.relplot(x="mpg", y="acceleration", col = 'model_year', data=df_car)
print('Tentativa 01 : Visualização entre gráficos diferentes não possível')
```

```python
plt.figure(figsize=(10,10))
plt.title('Relação entre Aceleração, Consumo e Ano de Fabricação')
sns.scatterplot(x="mpg", y="acceleration", hue = 'model_year', data=df_car)
plt.show()

print("Apesar da correlação fraca positiva considerando que a aceleração interfere no consumo de combustível, podemos perceber pela escurecimento gradual das cores que de fato os carros mais recentes possuem um menor consumo (Indicando uma correlação positiva, mas ainda fraca, entre o consumo e o ano de fabricação). Possivelmente, motivos pela Crise do Petróleo de 1973 que, em função da inflação dos preços da gasolina, acabou consolidando uma demanda de redução no peso dos automóveis e introduzindo novos materiais aos automóveis (Ex : Polímeros)")
```
