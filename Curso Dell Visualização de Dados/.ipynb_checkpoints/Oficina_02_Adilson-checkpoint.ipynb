```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_raw = "https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt"
df = pd.read_json(df_raw)
df.sort_values(by='Year',ascending=True)
```

##### Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
df_budget = df[["Film","Budget_M"]]

max_budget_filter = df_budget['Budget_M'].max()
min_budget_filter = df_budget['Budget_M'].min()

max_budget = df_budget[df_budget['Budget_M'] == max_budget_filter]['Film'].values
min_budget = df_budget[df_budget['Budget_M'] == min_budget_filter]['Film'].values


plt.figure(figsize=(10,10))
plt.title('Lista Distribuição Orçamento Filmes')
sns.set(style='darkgrid')
sns.barplot(y="Film",x='Budget_M',data=df_budget)
plt.show()

print(f'O filme de maior orçamento foi o : {max_budget}')
print(f'O filme de menor orçamento foi o : {min_budget}')
```

##### Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
df_Gross = df[["Film","Worldwide_Gross_M"]]

max_gross_filter = df_Gross['Worldwide_Gross_M'].max()
min_gross_filter = df_Gross['Worldwide_Gross_M'].min()

max_gross = df_Gross[df_Gross['Worldwide_Gross_M'] == max_gross_filter]['Film'].values
min_gross = df_Gross[df_Gross['Worldwide_Gross_M'] == min_gross_filter]['Film'].values


plt.figure(figsize=(10,10))
plt.title('Lista Distribuição Bilheteria Filmes')
sns.set(style='ticks')
sns.barplot(y="Film",x='Worldwide_Gross_M',data=df_Gross)
plt.show()

print(f'O filme de maior bilheteria foi o : {max_gross}')
print(f'O filme de menor bilheteria foi o : {min_gross}')
```

##### Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
df_category = df[['Genre','Year']]
df_category.groupby('Genre').count()
```

```python
df_Category_raw = df[["Film","Year","Genre","Budget_M"]]

plt.figure(figsize=(20,9))
plt.title('Distribuição Anual de Filmes por Gênero')
sns.barplot(x='Year',y='Budget_M',estimator=len,ci=None,hue='Genre',palette='Paired',data=df_Category_raw)
plt.ylabel('Quantidade de Filmes')
plt.xlabel('Ano')
plt.show()

# Considerando, para fins de exercício, uma análise gráfica :

print('Em que anos foram feitos menos filmes de ação?')
print('Os anos com menos filmes de ação foram 2008 e 2010')

print('\nEm que ano foram feitos menos filmes de animação')
print('Os anos com menos filmes de animação foram 2008 e 2011')
```

##### Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

```python
from numpy import mean

df_Category = df[["Film","Year","Genre","Budget_M"]]

plt.figure(figsize=(20,9))
plt.title('Distribuição Anual da Média de Orçamentos')
sns.barplot(x='Year',y='Budget_M',estimator=mean,ci=None,hue='Genre',palette='Paired',data=df_Category)
plt.ylabel('Média do Custo dos Orçamentos')
plt.xlabel('Ano')
plt.show()

# Considerando, para fins de exercício, uma análise gráfica :

print('Em que ano os filmes de ação tiveram maior média de orçamento?')
print('O ano em que os filmes de ação tiveram a maior média de orçamento foi 2007')

print('\nEm que ano os filmes de animação tiveram a maior média de orçamento?')
print('O ano em que os filmes de animação tiveram a maior média de orçamento foi 2008')
```
