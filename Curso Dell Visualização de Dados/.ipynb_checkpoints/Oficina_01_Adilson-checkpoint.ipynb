```python
import pandas as pd
df = pd.read_csv('drinks.txt')
```

##### Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?

```python
# Transformar Coluna em Lista
# df.describe() para validar os valores 

list_df = list(df.columns)

for column in list_df:
    if column != 'country':
        print(f"A média para {column} é de {df[column].mean()}")
    else:
        pass

print(" ")
    
for column in list_df:
    if column != 'country':
        print(f"O desvio Padrão para {column} é de {df[column].std()}")
    else:
        pass
```

#####  Qual o consumo de licor na França?

```python
df_France = df[df['country'] == 'France']
print(f"O consumo de licor na França é de {df_France['spirit_servings'].values}")

df_licor = df[['country','spirit_servings']]
df_licor.sort_values(by = 'spirit_servings', ascending=False).head(25)
df_licor.describe()

print(f"\nA média de consumo de licor é de {df_licor.mean().values} e o valor do Terceiro Quadrante é de {df_licor.quantile(0.75).values}")

print("\nAnalisando o consumido em todos os países, apesar da França estar acima da média e acima do terceiro quadrante, ainda assim quem domina o ranking de consumo de licor são países orientais, provavelmente por questões culturais.")

```

#####  Quais os 5 países nos quais se consome mais vinho?

```python
df_top5 = df[['country','wine_servings']]

print('Diferente da Questão de Licor, quem domina o ranking de consumo de vinho são os países europeus')

df_top5.sort_values(by = 'wine_servings', ascending=False).head(5)
```

#####  Quais os países nos quais o consumo de cerveja está acima da média?

```python
df_beer = df[['country','beer_servings']]
media_cons_beer_array = df_beer.mean().values
media_cons_beer = media_cons_beer_array[0]

df_beer[df_beer['beer_servings'] > media_cons_beer].sort_values(by='beer_servings',ascending=False)
```

#####  Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?

```python
filter_Brazil = df[df['country'] == 'Brazil']['total_litres_of_pure_alcohol'].values
filter_Brazil[0]

df_total_litres = df[df['total_litres_of_pure_alcohol'] > filter_Brazil[0]]
df_total_litres.sort_values(by='total_litres_of_pure_alcohol',ascending=True)
```

#### AVALIAÇÃO

```python
total_country = df['country'].count()
pior_que_brazil = df_total_litres['country'].count()

print(f'De um total de {total_country} países, temos {pior_que_brazil} países com o consumo em litros de alcool puro maior que o Brasil, ainda assim, considerando que estamos na posição de {pior_que_brazil+1}, temos que ficar alarmados com o risco de doenças e distúrbios ligados ao alcool no país, tendo em vista uma média mundial de {df["total_litres_of_pure_alcohol"].mean()}, menor que a nossa de 7.2')
```
