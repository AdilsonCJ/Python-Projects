```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="dark")
```

```python
df_cars = sns.load_dataset("mpg")
df_cars.head(3)
```

```python
sns.pairplot(df_cars,kind='reg',hue='origin')

print('O Dataset em questão é bem interessante pois nos mostra a correlação de variáveis no carro : Como potência, consumo, aceleração e torque')
print('Para as variáveis correlacionadas, podemos citar um grau alto de correlação : mpg com displacement, mpg com horsepower, mpg com weight, displacement com acceleration, displacement com horsepower, displacement com weight ... Ou seja todas as variáveis que estão ligadas com a performance do carro (Aceleração, Peso, Torque) e seu consumo (mpg) tem um grau de correlação')
```

###### Escolha uma distribuição conjunta em que as variáveis estejam fortemente correlacionadas e crie um gráfico que permita analisar a distribuição conjunta, a reta de regressão e as distribuições marginais.

```python
print('Optando por escolher a relação entre horsepower e weight')

sns.jointplot(x='horsepower',y='weight',kind='reg',data=df_cars)
```

```python
print('Apesar da média do horsepower ser 100, curiosamente vemos um pico em 150. Fazendo uma rápida análise e correlacionamento com a história do automobilismo e paixão dos Americanos por "caminhonetes Ford F250", podemos evidenciar que de fato o país consome muitas caminhonetes de alto torque, sendo os responsáveis pelo pico na frequência do histograma na casa dos 150 ')

df_horsepower = df_cars[df_cars['horsepower'] > 150]
df_horsepower.groupby(['origin']).count()
```
